# **subplot/subplots 를 이용한 여러개의 차트 그리기**

```python
import matplotlib.pyplot as plt
```

\- 문제 : data1(x:1,2,3;y:1,2,3)과 data2(x:1,2,3;y:1,100,200)을 그래프로 출력하시오

```python
x1 = (1,2,3)
y1 = (1,2,3)
x2 = (1,2,3)
y2 = (1,100,200)
```

- subplot을 이용한 해결
subplot(행의 수, 열의 수, 해당 그래프가 그려질 위치)

```python
plt.subplot(1,2,1)
plt.plot(x1, y1)
plt.title('data1')

plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.title('data2')
```

```python
plt.subplot(2,1,1)
plt.plot(x1,y1)
plt.title('data1')

plt.subplot(2,1,2)
plt.plot(x2,y2)
plt.title('data2')
# data1번의 x축과 data2의 title이 겹친다. 그럴때는 tight_layout() 사용
```

\- tight_layout을 이용하면 레이아웃이 자동으로 설정됨

```python
plt.subplot(2,1,1)
plt.plot(x1, y1)
plt.title('data1')

plt.subplot(2,1,2)
plt.plot(x2,y2)
plt.title('data2')
plt.tight_layout()
```

\- 문제 data1(x:1,2,3;y:1,2,3)과 data2(x:1,2,3;y:1,50,200)을 그래프로 출력하시오

```python
x1 = (1,2,3)
y1 = (1,2,3)
x1= (1,2,3)
y2 = 1,50,200
```

```python
fig, axis = plt.subplots()

axis2 = axis.twinx()
axis.plot(x1,y1, color='red', label='data1')
axis2.plot(x2,y2, color='blue', label='data2')

axis.set_xlabel('X', fontsize='15')
axis.set_ylabel('y1', fontsize='15')
axis2.set_ylabel('y2', fontsize='15')
```

- y 축을 두 개 가진 차트에서 범례 표시하기

```python
fig, axes =plt.subplots()

axes2 =axes.twinx()
chart1 = axes.plot(x1,y1, color = 'red', label='data1')
chart2 = axes2.plot(x2,y2,color='blue', label='data2')

axes.set_xlabel('X', fontsize=15)
axes.set_ylabel('y1', fontsize=15)
axes2.set_ylabel('y2', fontsize=15)

chart = chart1 + chart2
plt.legend(chart,['chart1', 'chart2'])
```
